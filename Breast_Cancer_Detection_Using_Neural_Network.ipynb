```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

```python
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
breast_cancer_dataset
```

```python
data_frame = pd.DataFrame(data=breast_cancer_dataset['data'],columns=breast_cancer_dataset['feature_names'])
```

```python
data_frame['label'] = breast_cancer_dataset.target
```

```python
x = data_frame.drop(columns='label',axis=1)
y = data_frame['label']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

```python
# Standardized the data
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
x_train_std = scalar.fit_transform(x_train)
x_test_std = scalar.transform(x_test)
```

```python
x_train_std
```

```python
x_test_std
```

```python
# Building the Neural Network
# Importing tensorflow and keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# Setting up the layers of Neural Network
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (30,)),
    keras.layers.Dense(20, activation='relu'),
    keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# Compiling the Neural Network
model.compile(optimizer='adam',
              loss = 'sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# Traiing the Neural Network
history = model.fit(x_train,y_train,validation_split=0.1,epochs=30)
```

```python
# See Difference after scalarizing values
history = model.fit(x_train_std,y_train,validation_split=0.1,epochs=30)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='upper right')
plt.show()
```

```python
# Accuracy of the model on test Data
loss , accuracy = model.evaluate(x_test_std,y_test)
print('Accuracy',accuracy)
print('Loss',loss)
```

```python
# Model.predict() gives me the prediction probability of each class for that data point
# Converting the prediction probabilities to class labels

y_pred_label = [np.argmax(i) for i in y_pred]
print(y_pred_label)
```
